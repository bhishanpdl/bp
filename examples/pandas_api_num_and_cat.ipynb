# Numerical and Categorical Plots

```python
df = titanic
```

```python
df.bp.plot_num('age',print_=True,ms='seaborn-darkgrid')
```

```python
df.bp.plot_cat('pclass')
```

```python
df.bp.plot_num_num('age','fare',xticks1=range(0,90,10),
                   xticks2=range(0,600,50),rot=90,figsize=(12,12))
```
